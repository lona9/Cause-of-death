## Cause of Death for 20 to 30 years old in Chile

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
columnas = ["anho_def", "fecha_def", "sexo", "edad_tipo", "edad_cant", "cod_comuna_res", "comuna_res", "region_res", "diag1", "capitulo_diag1", "glosa_cap_diag1", "cod_grupo_diag1", "glosa_grupo_diag1", "cod_categoria_diag1", "glosa_categoria_diag1", "cod_subcategoria_diag1", "glosa_subcategoria_diag1", "diag2", "capitulo_diag2", "glosa_cap_diag2", "cod_grupo_diag2", "glosa_grupo_diag2", "codigo_categoria_diag2", "glosa_categoria_diag2", "cod_subcategoria_diag2", "glosa_subcategoria_diag2"]
df_original = pd.read_csv("DEFUNCIONES_FUENTE_DEIS_2016_2021_22042021.csv", encoding="latin-1", sep=";", names=columnas)
```

```python
df = df_original.copy()
```

Borrar diag2, solo específico para caídas?

```python
df.drop(list(df.filter(regex="diag2")), axis=1, inplace=True)
```

```python
df.info()
```

Filtrar por edades, solo en años y solo entre 20 a 30.

```python
df_20a30 = df.query("edad_tipo == 1")
```

```python
df_20a30 = df_20a30.query("edad_cant >= 20 & edad_cant <= 30")
```

```python
df_20a30.info()
```

## Mujeres de 20 a 30 años

```python
mujeres = df_20a30.query("sexo == 'Mujer'")
```

```python
muj16 = mujeres.query("anho_def == 2016")
mcausas16 = pd.Series(muj16.glosa_subcategoria_diag1.value_counts()[0:5])
mcausas16
```

```python
muj17 = mujeres.query("anho_def == 2017")
mcausas17 = pd.Series(muj17.glosa_subcategoria_diag1.value_counts()[0:5])
mcausas17
```

```python
muj18 = mujeres.query("anho_def == 2018")
mcausas18 = pd.Series(muj18.glosa_subcategoria_diag1.value_counts()[0:5])
mcausas18
```

```python
muj19 = mujeres.query("anho_def == 2019")
mcausas19 = pd.Series(muj19.glosa_subcategoria_diag1.value_counts()[0:5])
mcausas19
```

```python
muj20 = mujeres.query("anho_def == 2020")
mcausas20 = pd.Series(muj20.glosa_subcategoria_diag1.value_counts()[0:5])
mcausas20
```

```python
mcausas_labels = ("Asfixia", "Traumatismos múltiples", "Traumatismo intracraneal", "Parálisis cerebral", "Tumor del cuello uterino", "Asfixia", "Traumatismos múltiples", "Traumatismo intracraneal", "Tumor del cuello uterino", "Otras causas no definidas", "Asfixia", "Traumatismos múltiples", "Otras causas no definidas", "Traumatismo intracraneal", "Otros traumatismos", "Asfixia", "Traumatismos múltiples", "Traumatismo intracraneal", "Otras causas no definidas", "Tumor maligno del cuello uterino", "Asfixia", "COVID-19", "Traumatismos múltiples", "Traumatismo intracraneal", "Otras causas no definidas")
```

```python
ind = np.arange(len(mcausas16))
ticks = np.arange(len(mcausas_labels))
width = 0.8
```

```python
plt.figure(figsize=(20,7.5), dpi=100)
plot16 = plt.bar(ind, mcausas16, width, label = "2016")
plt.bar(ind + 5, mcausas17, width, label = "2017")
plt.bar(ind + 10, mcausas18, width, label = "2018")
plt.bar(ind + 15, mcausas19, width, label = "2019")
plt.bar(ind + 20, mcausas20, width, label = "2020")
legends = ["2016", "2017", "2018", "2019", "2020"]

plt.ylabel('Número de muertes', fontsize=18)
plt.xlabel('Causas de muerte', fontsize=18)
plt.title('Causas principales de muerte de mujeres entre 20 y 30 años', fontsize=24)
locations = ticks # xtick locations
labels = mcausas_labels
plt.xticks(locations, labels, rotation = 90, fontsize=14)
plt.yticks(fontsize=14)
plt.legend(legends, fontsize=14)
```
